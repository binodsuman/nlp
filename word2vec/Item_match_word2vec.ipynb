```python
import requests
from bs4 import BeautifulSoup
import gensim.downloader as api

# Load pre-trained word2vec model
model = api.load('word2vec-google-news-300')

```

```python
# Define a function to get text content from a webpage
def get_text(url):
    res = requests.get(url)
    html_page = res.content
    soup = BeautifulSoup(html_page, 'html.parser')
    text = soup.get_text().replace('\n', ' ')
    return text

# Sample competitor website urls
competitor_urls = ['https://www.amazon.com', 'https://www.walmart.com']

```

```python
# Sample query word
query = 'shoes'
text = get_text('https://www.amazon.com/s?k=shoe')
text
```

```python
competitor_texts = []
for url in competitor_urls:
    print("URL : ", url)
    text = get_text(url)
    print("Text: ", text)
    competitor_texts.append(text)

```

```python
# Get text content from each competitor website
competitor_texts = []
for url in competitor_urls:
    text = get_text(url)
    competitor_texts.append(text)

# Split text content into sentences
sentences = [sentence.strip().split() for text in competitor_texts for sentence in text.split('. ')]

# Find most similar words to the query word in the pre-trained model
similar_words = model.most_similar(query)

# Print the top 10 most similar words to the query word and their corresponding similarity scores
print('the top 10 most similar words to the query word and their corresponding similarity scores')
for word, score in similar_words[:10]:
    print(f'{word}: {score}')

# Find sentences that contain the most similar words to the query word
print('Find sentences that contain the most similar words to the query word')
matching_sentences = []
for sentence in sentences:
    for word, _ in similar_words:
        if word in sentence:
            matching_sentences.append(' '.join(sentence))
            break

# Print the top 10 matching sentences
print('the top 10 matching sentences')
for sentence in matching_sentences[:10]:
    print(sentence)


```
